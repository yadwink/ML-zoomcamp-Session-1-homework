# Question 5
1. We have 9 features: 7 numerical features and 2 categorical.
    Let's find the least useful one using the feature elimination technique.
2. Train a model with all these features (using the same parameters as in Q4).
    Now exclude each feature from this set and train a model without it. Record the accuracy for each model.
3. For each feature, calculate the difference between the original accuracy and the accuracy without the feature.
4. Which of following feature has the smallest difference?
    neighbourhood_group
    room_type
    number_of_reviews
    reviews_per_month
note: the difference doesn't have to be positive

```python
# Feature elimination technique
from IPython.display import display

for c in categorical:
    df_group = df_full_train.groupby(c).price.agg(['mean','count'])
    df_group['diff'] = df_group['mean'] - global_price
    df_group['risk'] = df_group['mean'] / global_price
    display(df_group)
    print()
    print()
```

```python
# make a function for all features and keep eliminating each feature when trainig models
def calculate_accuracy(features):
    
    # one-hot encoding datasets
    dv = DictVectorizer(sparse=False)

    train_dict = df_train[features].to_dict(orient='records')
    val_dict = df_val[features].to_dict(orient='records')

    X_train = dv.fit_transform(train_dict)
    X_val = dv.transform(val_dict)

    # trainign models
    model = LogisticRegression(solver='liblinear', C=1.0, random_state=42)
    model.fit(X_train, y_train)

    # apply models on vaöidation datasets
    y_pred = model.predict_proba(X_val)[:,1]

    #add a threshold
    decision = (y_pred >= 0.5)

    # accuracy
    accuracy = (y_val == decision).mean()
    

    df_pred = pd.DataFrame()
    df_pred['probability'] = y_pred
    df_pred['prediction'] = decision
    df_pred['actual'] = y_val
    df_pred['correct'] = df_pred.prediction == df_pred.actual
    return accuracy, df_pred

acc, df_pred = calculate_accuracy(numerical+categorical)    

print(acc)
df_pred.head()
```

```python
# store the accuracy of all variables
all_acc = 0.79
```

```python
all_features = numerical+categorical
feat_diff = {}

for f in all_features:
    features = all_features.copy()
    features.remove(f)
    accuracy,_ = calculate_accuracy(features)
    feat_diff[f] = all_acc - accuracy
    
feat_diff    
```

# Answer Q5:The smallest difference in accuracy is from reviews per month

1. For this question, we'll see how to use a linear regression model from Scikit-Learn
2. We'll need to use the original column 'price'. Apply the logarithmic transformation to this column.
3. Fit the Ridge regression model on the training data.
4. This model has a parameter alpha. Let's try the following values: [0, 0.01, 0.1, 1, 10]
5. Which of these alphas leads to the best RMSE on the validation set? Round your RMSE scores to 3 decimal digits.
6. If there are multiple options, select the smallest alpha.

```python
# linear regression with price feature included
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_full_train, test_size=0.2, random_state=42)
len(df_train), len(df_val), len(df_test)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
```

```python
# select all features with numerical and categorical data types seperately:
numerical = ['price','latitude', 'longitude', 'minimum_nights','number_of_reviews','reviews_per_month','calculated_host_listings_count','availability_365']
categorical = ['neighbourhood_group', 'room_type']
```

```python
#log transformation to the price
y_train = np.log1p(df_train['price'].values)
y_val = np.log1p(df_val['price'].values)
y_test = np.log1p(df_test['price'].values)
```

```python
# remove price
del df_train['price']
del df_val['price']
del df_test['price']
```

```python
dv = DictVectorizer(sparse=False)

train_dict = df_train[categorical].to_dict(orient='records')
val_dict = df_val[categorical].to_dict(orient='records')

X_train = dv.fit_transform(train_dict)
X_val = dv.transform(val_dict)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
scores = {}
for alpha in [0, 0.01, 0.1, 1, 10]:
    model = Ridge(alpha=alpha)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    score = np.sqrt(mean_squared_error(y_val, y_pred))
    scores[alpha] = score.round(3)
    print(alpha, score)
```
